#### On TASO graph

```python
graph = taso.new_graph()
input = graph.new_input(dims=(1,128,56,56))
w1 = graph.new_weight(dims=(128,128,3,3))
w2 = graph.new_weight(dims=(128,128,1,1))
w3 = graph.new_weight(dims=(128,128,3,3))
left = graph.conv2d(input=input, weight=w1, strides=(1,1), padding="SAME", activation="RELU")
left = graph.conv2d(input=left, weight=w3, strides=(1,1), padding="SAME")
right = graph.conv2d(input=input, weight=w2, strides=(1,1), padding="SAME", activation="RELU")
output = graph.add(left, right)
output = graph.relu(output)

print("cost = {}".format(graph.cost()))
new_graph = taso.optimize(graph, alpha = 1.0, budget = 100, print_subst=True)
print("\n optimized_cost = {}".format(new_graph.cost()))
```

#### On Torch graph

```python
class SampleModel(nn.Module):
    def __init__ (self):
        super().__init__()
        self.left_conv1 = nn.Conv2d(3,3,3,padding=1)
        self.relu = nn.ReLU()
        self.left_conv2 = nn.Conv2d(3,3,3,padding=1)
        self.right_conv1 = nn.Conv2d(3,1,1)
    def forward(self,x):
        l = self.left_conv1(x)
        l = self.relu(l)
        l = self.left_conv2(l)
        r = self.right_conv1(x)
        r= self.relu(r)
        x = l+r
        x = self.relu(x) 
        return x

model = SampleModel()
x = torch.randn(1,3,56,56)
model(x)
torch.onnx.export(model,
    x,
    "model.onnx",
    verbose=False,
)
graph = taso.load_onnx("./model.onnx")
print("\n cost = {}".format(graph.cost()))
new_graph = taso.optimize(graph, alpha = 1.0, budget = 100, print_subst=True)
print("\n optimized_cost = {}".format(new_graph.cost()))
new_model = taso.export_onnx(new_graph)
onnx.save(new_model, "./model_taso.onnx")
```

#### Sample Model 2

```python
class SampleModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3,3,3,padding=1)
        self.conv2 = nn.Conv2d(3,3,3,padding=1)
        self.relu = nn.ReLU()
        self.fcl1 = nn.Linear(1,14)
        self.fcl2 = nn.Linear(14,14)
        
    def forward(self,X):
        x = self.conv1(X)
        x = self.conv1(x)
        x = self.relu(x)
        
        y = self.conv2(X)
        y = self.conv2(y)
        y = self.relu(y)
        
        x = x+y
        x = x*2
        x = x*2
        x = x+1
        x = x-1
        
        return x

model = SampleModel()
x = torch.randn(1, 3, 24, 24, device='cpu')
torch.onnx.export(model,
    x,
    "model.onnx",
    verbose=False,)
graph = taso.load_onnx("./model.onnx")
print("\n cost = {}".format(graph.cost()))
new_graph = taso.optimize(graph, alpha = 1.0, budget = 100, print_subst=True)
print("\n optimized_cost = {}".format(new_graph.cost()))
new_model = taso.export_onnx(new_graph)
onnx.save(new_model, "./model_taso.onnx")
```
